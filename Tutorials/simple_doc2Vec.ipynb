```python
from gensim.models.doc2vec import Doc2Vec,\
    TaggedDocument
from nltk.tokenize import word_tokenize
```

```python
# define a list of documents.
data = ["Самая мощная вспышка за последние шесть лет зафиксирована на Солнце",
        "Сотрудники центра также предупредили, что подобные активные события на Солнце зачастую предвещают магнитные бури, ближайшая из них ожидается в субботу",
        "Банковские вклады, которые включают в себя инвестпродукты, могут оказаться совсем не выгодными с точки зрения дохода, рассказал агентству «Прайм» финансист Лазарь Бадалов.",
        "С 1 января изменятся правила предоставления компенсаций и субсидий. Более того, вырастет число их получателей, а также размер выплат. "]
```

```python
tagged_data = [TaggedDocument(words=word_tokenize(doc.lower()),
                              tags=[str(i)]) for i,
               doc in enumerate(data)]
```

```python
tagged_data
```

```python
# train the Doc2vec model
model = Doc2Vec(vector_size=20,
                min_count=2, epochs=50)
model.build_vocab(tagged_data)
model.train(tagged_data,
            total_examples=model.corpus_count,
            epochs=model.epochs)
```

```python
# get the document vectors
document_vectors = [model.infer_vector(
    word_tokenize(doc.lower())) for doc in data]
```

```python
#  print the document vectors
for i, doc in enumerate(data):
    print("Document", i+1, ":", doc)
    print("Vector:", document_vectors[i])
    print()
```
