```python
from plsa import Corpus, Pipeline, Visualize
from plsa.pipeline import DEFAULT_PIPELINE
from plsa.algorithms import PLSA
```

```python
csv_file = 'RPD_output/dataset_res.csv'
```

```python
PIPELINE = (
    tokenize,
    LemmatizeWords('NN'),
)
```

```python
pipeline = Pipeline(*PIPELINE)
pipeline
```

```python
corpus = Corpus.from_csv(csv_file, pipeline, encoding = "utf-8")
corpus
```

```python
n_topics = 46
```

```python
plsa = PLSA(corpus, n_topics, True)
plsa
```

```python
result = plsa.fit()
plsa
```

```python
result = plsa.best_of(5)
```
